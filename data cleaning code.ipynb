```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('synthetic_online_retail_data.csv')

df.head()
```

```python
print(df.shape)

df.info()

df.isnull().sum()
```

```python
df['review_score'].fillna(df['review_score'].median(), inplace=True)
```

```python
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)

print("Duplicates Removed")
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])

df.info()
```

```python
df.to_csv('cleaned_online_retail_data.csv', index=False)

print("Data Cleaning Completed")
```
